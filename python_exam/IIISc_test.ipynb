```python
input_Q1Q2 = [(253, 127), (139, 305), (148, 99), (363, 134), (778, 396), (650, 759), (724, 547), (788, 412), (105, 1)]
    
output_Q1Q2 = [38, 59, 29, 76, 30, 70, 53, 59, 51]

result =[]

for item in input_Q1Q2:
    source = item[0]
    destination = item[1]
    shortest_dist = -1

    
    # connections holds the graph. It is a dictionary of dictionaries that stores the neighbor nodes and their 
    # costs for each node
    connections = {}

    for node in range(1,max(graph_object["init_node"])+1):
        connections[node] = {}    

    for i in range(graph_object.shape[0]):
        row = graph_object.iloc[i]
        start = row.init_node
        end = row.term_node
        cost = row.free_flow_time 
        connections[start][int(end)] = cost
        
     
    # creating a list of min cost to reach each node from source
    mincost = [sys.maxsize for i in range(max(graph_object["init_node"]))]
    mincost[source] = 0

        
    # creating a dictionary of neighbor node and its cost. the node with min cost is popped 
    # from here and used to find neighbors from that
    node_pq = {}
    node_pq[source] = 0

    while len(node_pq) > 0:

        node_min = min(node_pq, key=lambda x: node_pq[x])      # return node with min cost from the dict
        node_min_cost = node_pq[node_min]

        for neighbor,cost in connections[node_min].items():
            if node_min_cost + cost < mincost[neighbor-1]:
                mincost[neighbor-1] = node_min_cost + cost
                node_pq[neighbor] = mincost[neighbor-1]

        del node_pq[node_min]

        
        
    if(mincost[destination] != sys.maxsize):
        shortest_dist = mincost[destination-1]
        
    result.append(shortest_dist)
    
    
    
result
```
